##   PlaceHolders : 
A placeholder is simply a variable that we will assign data to at a later date. It allows us to create our operations and build our computation graph, without needing the data. In TensorFlow terminology, we then feed data into the graph through these placeholders.

## Variables :
Variables are manipulated via the tf.Variable class. A tf.Variable represents a tensor whose value can be changed by running operations on it.Internally, a tf.Variable stores a persistent tensor. Specific operations allow you to read and modify the values of this tensor. These modifications are visible across multiple tf.Sessions, so multiple workers can see the same values for a tf.Variable.
(Tensor : A high dimensional array)
### Source : https://www.tensorflow.org/guide/variables


### So lets make a Tensorflow PlaceHolder for target and features!
### And make Tensorflow Variables for Slopes and intercepts! 



```
# tf.placeholder(type(numbers to consider), shape(matrix))
# Here None represents that, later when data is fed, then None moulds(changes) according to the given data.
target_PH = tf.placeholder(tf.float64, [None,10])
features_PH = tf.placeholder(tf.float64, [None,784])
```

```
# Making Tensorflow Variables for Slopes(Weights) & intercepts(Bias)
m1 = np.random.randn(784,784)
m1 = tf.Variable(m1)
c1 = random.random()
c1 = tf.cast(c1, tf.float64) # Since C values are float32 , and tensors accept only float 64, we're converting all c values to float64
c1 = tf.Variable(c1)

m2 = np.random.randn(784,784)
m2 = tf.Variable(m2)
c2 = random.random()
c2 = tf.cast(c2, tf.float64)
c2 = tf.Variable(c2)

m3 = np.random.randn(784,784)
m3 = tf.Variable(m3)
c3 = random.random()
c3 = tf.cast(c3, tf.float64)
c3 = tf.Variable(c3)

m4 = np.random.randn(784,784)
m4 = tf.Variable(m4)
c4 = random.random()
c4 = tf.cast(c4, tf.float64)
c4 = tf.Variable(c4)

m5 = np.random.randn(784,784)
m5 = tf.Variable(m5)
c5 = random.random()
c5 = tf.cast(c5, tf.float64)
c5 = tf.Variable(c5)

m6 = np.random.randn(784,10)   # Why (784, 10) will be explained soon.
m6 = tf.Variable(m6)
c6 = random.random()
c6 = tf.cast(c6, tf.float64)
c6 = tf.Variable(c6)
# Reason to make 6m's and 6c's, will be explained soon.
```

# Now our goal is to make Layers!
Our Hypothesis Function is **Y = mx + c**.
Where, 

*   m = slope (weight)
*   c = intercept (Bias)

---


Now our Y will fetch us predicted value of that particular x with some given m & c right.
That was our Traditional ****Machine Learning****.


---

#### Now in **Deep Learning** which involves the study of **Neural Networks**, we make Function of Functions such as **Y = mx +c** was our Hypothesis function right, we make another Hypothesis function on top of this and on top of whole of that and so on.... which forms **Layers**.
And that is what is going to fetch us Accurate Predictions with **optimal m & c**.




## Now, Let's break down the problem of MNIST classification :
In MNIST dataset there are,


*   20,000(20k) images and,
*   784 features for each image


---

Our Hypothesis Function to classify each image is going to be **Y=m.X + c**.

---

Steps to Classify each image : 

1.   Make 784 **Neurons** which are nothing but denotes each **feature** of the inputted image whose shape is (**N,784**) for $ N^{th} $ image. (**X**)

2.   Make a slopes array(**m1**) whose shape is **(784,784)** and dot product it with **X** and add an intercept value (**c**). The output array is going to be (**N,784**).
### Y = X . m+c
### (N,784) = (N, 784) . (784,784) + c

3.  Now store the above outputted matrix into a variable **J**. And dot product it with another slope matrix of shape (**784,784**) and store the ouputted matrix into another new variable **K**. 
4. Like this make as many matrices as u wish.
5. For the final layer, the slopes matrix is going to turn (**784,1**) and the final layer when dot product with this slopes matrix, its going to fetch us with a single value. Hence that is our **Y predicted** value of that inputted image.











```
L1Out = tf.matmul(features_PH, m1)+c1  #Denotes i in the image
L2Out = tf.matmul(L1Out, m2)+c2        #Denotes j in the image
L3Out = tf.matmul(L2Out, m3)+c3        #Denotes k in the image
L4Out = tf.matmul(L3Out, m4)+c4        #Denotes l in the image
L5Out = tf.matmul(L4Out, m5)+c5        #Denotes F in the image
L6Out = tf.matmul(L5Out, m6)+c6        #Denotes O in the image
```

# OneHot Encoding
Now we need to do something called One-Hot Encoding to our final output!


---
One-hot encodings transform our categorical labels into vectors of 0s and 1s. The length of these vectors is the number of classes or categories that our model is expected to classify.
### Source : https://deeplizard.com/learn/video/v_4KWmkwmsU


---
Below Image, shows an example of how OneHot encoding takes place :


### As u can see, number denoting 4 image sets "1" in the position 4, similarly various other images sets 1 in their respective positions.
In the below image, when m6 = (784,10) and when this gets dot product with final layer(F), it returns us (1,10) and the final output value for image "4" will be stored into the position of "4" in the Output array. The value stored in the position "4" need not be 1, it can be any value depending upon weights that are optimised and "SV" in the output array denotes that, remaining all values need not be "0" but any value.
#### Edit : In the below image all the arrows should point at 4th neuron of output array.


```
# Making a Function to make a OneHot array
def OneHot(x):
  zeros = [0]*10
  for i in range(len(zeros)):
    zeros[x] = 1;
  return zeros
OneHot(9)
```

```
#OneHot encoding all Actual values(i.e) our target values
target_OH = []
for i in range(len(target)):
  target_OH.append(OneHot(target.iloc[i]))
target_OH = np.array(target_OH)
```

```
target_OH
```

# Computing Cost

Cost is computed with using regular Mean Squared Error function.




1.   Setting up cost
2.   Setting up Optimization using Gradient Descent optimizer
3. Minimising cost by setting up cost inside optimizer



```
cost = tf.reduce_mean((L6Out - target_PH)**2)
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.00000000001)   #Setting up Gradient Descent optimizer
```

```
minimizer = optimizer.minimize(cost)  # Minimising cost by setting up cost inside optimizer
```

### In the below code, there are 3 important lines to notice :


1.  **initialisation** : Place where something is considered as global_variables and those global_variables will be initialised.
2.   **Session**: Place where a "Specified" session takes place.
3. ***Session.run(initialisation)*** : This is how Session of "initialisation" takes place.



```
initialisation = tf.global_variables_initializer()
session = tf.Session()
session.run(initialisation)
```

### In the below code, we're making dictionary and that dictionary consists of features and target variables and those are going to be used by the tensorflow.

```
Dict = {features_PH : np.array(features),target_PH : np.array(target_OH)}
```

```
#from IPython.display import clear_output
it = 1000
error = []
for i in range(it):
  session.run(minimizer, feed_dict = Dict)
  error.append(session.run(cost,feed_dict = Dict))
  #clear_output(True)
  print("Iteration : ",i,"error :",session.run(cost,feed_dict = Dict))
```
